```python
df1 = df.copy()
```

```python
def update_job_tile(dataframe,original_title, updated_title):
    string_df = dataframe.job_title.str.contains(original_title)
    df1.loc[string_df, 'Updated_Job_Title'] = updated_title
```

```python
counts = df1['employee_residence'].value_counts()
data = {'Residence':counts.index,'Original_counts':counts.values,'Normalized_values':df1['employee_residence'].value_counts(normalize=True).values} 
```

```python
df_normalised = pd.DataFrame(data)  
df_normalised
```

#### Insights
- **With normalize set to True , returns the relative frequency by dividing all values by the sum of values.** 
- In employee residence column where majority of the values have very less frequence.
- We can change those values to 'others' where normalized value is less than 0.03.

```python
df1.loc[df1['employee_residence'].map(df1['employee_residence'].value_counts(normalize=True).lt(0.03)), 'employee_residence'] = 'Others'
```

```python
df1.employee_residence.value_counts()
```

```python
counts = df1['company_location'].value_counts()
data = {'Residence':counts.index,'Original_counts':counts.values,'Normalized_values':df1['company_location'].value_counts(normalize=True).values} 
```

```python
df_normalised = pd.DataFrame(data)  
df_normalised
```
